## Introduction to Tensors

### Creating tensors

```python
# scalar
scalar = torch.tensor(7)
scalar
```

```python
scalar.ndim
```

```python
scalar.item()
print(type(scalar.item()))
print(type(scalar))
```

```python
# Vector
vector = torch.tensor([7, 7])
vector
```

```python
vector.ndim
```

```python
print(vector.shape)
print(type(vector.shape))
print(vector.shape[0], type(vector.shape[0]))
```

```python
# Matrix
MATRIX = torch.tensor([[7, 8], [9, 10]])
MATRIX
```

```python
MATRIX.ndim
```

```python
print(MATRIX.shape)
print(type(MATRIX.shape))
print(MATRIX.shape[0], MATRIX.shape[1], type(MATRIX.shape[0]))
```

```python
# TENSOR
TENSOR = torch.tensor([[[1, 2, 3], [4,6,7], [8,9,10]], [[11, 12, 13], [14,16,17], [18,19,20]]])
TENSOR
```

```python
TENSOR.ndim
```

```python
TENSOR.shape
```

```python
TENSOR[0]
```

```python
TENSOR[1]
```

### Random Tensors

```python
# Create a random tensor of shape (3,4)
randomTensor = torch.rand(3, 3, 4)
randomTensor
```

```python
randomTensor.ndim
```

```python
randomTensor.shape
```

```python
# height, width, 3 color channels
random_image_tensor = torch.rand(size=(224,224,3))
random_image_tensor.shape, random_image_tensor.ndim
```

```python
plt.imshow(random_image_tensor)
```

### Zeros and Ones

```python
zeros = torch.zeros(size=(3,4))
zeros
```

```python
zeros * torch.rand(size=(3,4))
```

```python
# Create a tensor of all ones
ones = torch.ones(size=(3, 4))
ones
```
